```python
df['날짜'] = pd.to_datetime(df['날짜'].astype(str).str.replace('.', '-'), format='%Y-%m')

# 정책 시행 시점 설정 (2015년 6월)
policy_date_1 = '2017-09'
policy_date_2 = '2020-06'

# 날짜 열을 datetime 형식으로 변환
policy_date_1 = pd.to_datetime(policy_date_1.replace('.', '-'), format='%Y-%m')
policy_date_2 = pd.to_datetime(policy_date_2.replace('.', '-'), format='%Y-%m')

# 처치 그룹 표시
df['period'] = np.select(
    [df['날짜'] < policy_date_1, 
     (df['날짜'] >= policy_date_1) & (df['날짜'] < policy_date_2), 
     df['날짜'] >= policy_date_2],
    ['pre', 'mid', 'post'],
    default='pre'
)

# 더미 변수 생성
df['mid_period'] = (df['period'] == 'mid').astype(int)
df['post_period'] = (df['period'] == 'post').astype(int)

# 이중차분법 분석
X = sm.add_constant(df[['mid_period', 'post_period']])
y = df['아파트매매가격']

model = sm.OLS(y, X).fit()

print(model.summary())

# 각 기간의 평균 가격 계산
pre_policy = df[df['period'] == 'pre']['아파트매매가격'].mean()
mid_policy = df[df['period'] == 'mid']['아파트매매가격'].mean()
post_policy = df[df['period'] == 'post']['아파트매매가격'].mean()

print(f"\n첫 번째 정책 시행 전 평균 가격: {pre_policy:.2f}")
print(f"두 정책 사이 평균 가격: {mid_policy:.2f}")
print(f"두 번째 정책 시행 후 평균 가격: {post_policy:.2f}")
print(f"첫 번째 정책 후 가격 변화: {mid_policy - pre_policy:.2f}")
print(f"첫 번째 정책 후 가격 증감률: {mid_policy*100/pre_policy:.2f}%")
print(f"두 번째 정책 후 가격 변화: {post_policy - mid_policy:.2f}")
print(f"두 번째 정책 후 가격 증감률: {post_policy*100/mid_policy:.2f}%")
print(f"전체 가격 변화: {post_policy - pre_policy:.2f}")
```
